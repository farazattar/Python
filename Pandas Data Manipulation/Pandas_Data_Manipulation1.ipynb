# Removing Samples with Null Values in Dataset

```python
import pandas as pd
```

```python
# read the dataset
data = pd.read_csv('bigmart_data.csv')
```

```python
# view the top results
data.head()
```

```python
data.describe()
```

```python
data.isnull()
```

```python
#count the null values
data.isnull().sum()
```

```python
# drop the null values
data_without_null_values = data.dropna(how="any")
```

```python
data_without_null_values.isnull().sum()
```

# Counting the number of removed samples

```python
data_index = data.index
data_number_of_rows = len(data_index)
print(data_number_of_rows)
```

```python
data_without_null_values_index = data_without_null_values.index
data_without_null_values_number_of_rows = len(data_without_null_values_index)
print(data_without_null_values_number_of_rows)
```

```python
number_of_removed_samples = data_number_of_rows - data_without_null_values_number_of_rows
print(number_of_removed_samples )
```
