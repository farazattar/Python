# Combining Dataframes using concat()- Column-wise

```python
import pandas as pd
```

```python
# Read the dataset
data=pd.read_csv('bigmart_data.csv')
```

```python
# Counting the number of rows and columns of the dataframe.
data_rows = len(data.index)
data_columns = len(data.columns)
data_rows, data_columns
```

```python
data.head()
```

```python
# Splitting the data Column-wise
data_part1 = data[['Item_Identifier','Item_Weight']] 
data_part1
```

```python
data_part2 = data['Item_Fat_Content']
data_part2
```

```python
data_part1_rows = len(data_part1.index)
data_part1_columns = len(data_part1.columns)
data_part1_rows, data_part1_columns
```

```python
type(data_part2)
```

```python
# Merging dataframes with concat() column-wise.
# Axis 1 represents columns.
combined_data = pd.concat([data_part1, data_part2], axis=1)
combined_data
```

```python
combined_data_rows = len(combined_data.index)
combined_data_columns = len(combined_data.columns)
combined_data_rows, combined_data_columns
```

```python
combined_data.iloc[0, 2]
```

```python
# Splitting the data Column-wise using iloc
data_part3 = data.iloc[:,:5] 
data_part3
```

```python
data_part4 = data.iloc[:,5:] 
data_part4
```

```python
data_part3_rows = len(data_part3.index)
data_part3_columns = len(data_part3.columns)
data_part3_rows, data_part3_columns
```

```python
data_part4_rows = len(data_part4.index)
data_part4_columns = len(data_part4.columns)
data_part4_rows, data_part4_columns
```

```python
# Merging dataframes with concat() column-wise.
# Axis 1 represents columns.
combined_data2 = pd.concat([data_part3, data_part4], axis=1)
combined_data2
```

```python
combined_data2_rows = len(combined_data2.index)
combined_data2_columns = len(combined_data2.columns)
combined_data2_rows, combined_data2_columns
```
