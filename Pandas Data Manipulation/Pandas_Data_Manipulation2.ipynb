# Combining Dataframes using concat()

```python
import pandas as pd
```

```python
# Read the dataset
data=pd.read_csv('bigmart_data.csv')
```

```python
# Counting the number of rows and columns of the dataframe.
data_rows = len(data.index)
data_columns = len(data.columns)
data_rows, data_columns
```

```python
# Splitting the data
data_part1 = data[:3000]
data_part2 = data[3000:]
```

```python
data_part1.tail()
```

```python
data_part2.head()
```

```python
data_part1_rows = len(data_part1.index)
data_part1_columns = len(data_part1.columns)
data_part1_rows, data_part1_columns
```

```python
data_part2_rows = len(data_part2.index)
data_part2_columns = len(data_part2.columns)
data_part2_rows, data_part2_columns
```

```python
# Merging dataframes with concat(), note that the default is: join='outer' and axis=0
# axis 0 represents rows and axis 1 represents columns.
# join_axes is Deprecated since version 0.25.0.
combined_data = pd.concat([data_part1, data_part2], keys = ['part1', 'part2'])
combined_data
```

```python
combined_data_rows = len(combined_data.index)
combined_data_columns = len(combined_data.columns)
combined_data_rows, combined_data_columns
```

```python
# Retrieving part one of the combined data using 'loc' method. 
combined_data.loc['part1']
```

```python
combined_data_part1_rows = len(combined_data.loc['part1'].index)
combined_data_part1_columns = len(combined_data.loc['part1'].columns)
combined_data_part1_rows, combined_data_part1_columns
```
