# Dimentionality Reduction No. 7: TSNE and UMAP , Moon Data set

```python
# Import the required libraries.
import pandas as pd
import numpy as np
import umap
from sklearn.manifold import TSNE
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt
```

```python
# Make the moon data set.
X, y = make_moons(n_samples=1000, noise=0.01)
```

```python
# Show input data (X).
X
```

```python
# Determine the type of X.
type(X)
```

```python
# Determine the size of X.
X.shape
```

```python
# Show output data (y).
y
```

```python
# Plot the scatter of the data.
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
# Create a new TSNE.
# Reduce two dimension to one dimension (n_component=1).
# Note that we specify the random state.
tsne = TSNE(n_components=1, perplexity=50 ,random_state=25)
```

```python
# Learn from the data and transform the reducted data into a new array.
tsne_result = tsne.fit_transform(X)
```

```python
# Show the result.
tsne_result
```

```python
# Plot the scatter of the result.
# Note that the result is in one dimension.
# Only for sake of clarity, we add a supreficial dimension with this command:
# range(0, len(pca_result)).
plt.scatter(tsne_result[:, 0], range(0, len(tsne_result)), c=y)
```

```python
# Create a new UMAP.
# Reduce two dimension to one dimension (n_component=1).
umap_dr = umap.UMAP(n_components=1)
```

```python
# Learn from the data.
umap_dr.fit(X)
```

```python
# Transform the reducted data into a new array.
umap_result = umap_dr.transform(X)
```

```python
# Show the result.
umap_result
```

```python
# Plot the scatter of the result.
# Note that the result is in one dimension.
# Only for sake of clarity, we add a supreficial dimension with this command:
# range(0, len(pca_result)).
plt.scatter(umap_result[:, 0], range(0, len(umap_result)), c=y)
```
