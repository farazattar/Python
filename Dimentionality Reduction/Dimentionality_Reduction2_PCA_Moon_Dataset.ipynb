# Dimentionality Reduction No. 2: PCA , Moon Data set

```python
# Import the required libraries.
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA, KernelPCA
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt
```

```python
# Make the moon data set.
X, y = make_moons(n_samples=1000, noise=0.01)
```

```python
# Show input data (X).
X
```

```python
# Determine the type of X.
type(X)
```

```python
# Determine the size of X.
X.shape
```

```python
# Show output data (y).
y
```

```python
# Plot the scatter of the data.
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
# Create a new PCA.
# Reduce two dimension to one dimension (n_component=1).
# Note that we specify the random state.
pca = PCA(n_components=1, random_state=25)
```

```python
# Learn from the data.
pca.fit(X)
```

```python
# Transform the reducted data into a new array.
pca_result = pca.transform(X)
```

```python
# Show the result.
pca_result
```

```python
# Plot the scatter of the result.
# Note that the result is in one dimension.
# Only for sake of clarity, we add a supreficial dimension with this command:
# range(0, len(pca_result)).
plt.scatter(pca_result[:, 0], range(0, len(pca_result)), c=y)
```

```python
# Create a new kernel PCA.
kernel_pca = KernelPCA(n_components=1, kernel='poly', gamma=10, random_state=25)
```

```python
# Learn from the data.
kernel_pca.fit(X)
```

```python
# Transform the reducted data into a new array.
kernel_pca_result = kernel_pca.transform(X)
```

```python
# Show the result.
kernel_pca_result
```

```python
# Plot the scatter of the result.
# Note that the result is in one dimension.
# Only for sake of clarity, we add a supreficial dimension with this command:
# range(0, len(kernel_pca_result)).
# The kernel PCA result seems better than the PCA result.
plt.scatter(kernel_pca_result[:, 0], range(0, len(kernel_pca_result)), c=y)
```
