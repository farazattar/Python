# Dimentionality Reduction No. 6: ICA and ISOMAP , Moon Data set

```python
# Import the required libraries.
import pandas as pd
import numpy as np
from sklearn.decomposition import FastICA
from sklearn.manifold import Isomap
from sklearn.datasets import make_moons
from matplotlib import pyplot as plt
```

```python
# Make the moon data set.
X, y = make_moons(n_samples=1000, noise=0.01)
```

```python
# Show input data (X).
X
```

```python
# Determine the type of X.
type(X)
```

```python
# Determine the size of X.
X.shape
```

```python
# Show output data (y).
y
```

```python
# Plot the scatter of the data.
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
# Create a new ICA.
# Reduce two dimension to one dimension (n_component=1).
# Note that we specify the random state.
ica = FastICA(n_components=1, random_state=25)
```

```python
# Learn from the data.
ica.fit(X)
```

```python
# Transform the reducted data into a new array.
ica_result = ica.transform(X)
```

```python
# Show the result.
ica_result
```

```python
# Plot the scatter of the result.
# Note that the result is in one dimension.
# Only for sake of clarity, we add a supreficial dimension with this command:
# range(0, len(pca_result)).
plt.scatter(ica_result[:, 0], range(0, len(ica_result)), c=y)
```

```python
# Create a new ISOMAP.
# Reduce two dimension to one dimension (n_component=1).
isomap = Isomap(n_components=1)
```

```python
# Learn from the data.
isomap.fit(X)
```

```python
# Transform the reducted data into a new array.
isomap_result = isomap.transform(X)
```

```python
# Show the result.
isomap_result
```

```python
# Plot the scatter of the result.
# Note that the result is in one dimension.
# Only for sake of clarity, we add a supreficial dimension with this command:
# range(0, len(pca_result)).
plt.scatter(isomap_result[:, 0], range(0, len(isomap_result)), c=y)
```
